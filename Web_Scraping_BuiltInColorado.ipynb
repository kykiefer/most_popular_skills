```python
from bs4 import BeautifulSoup # For HTML parsing
import urllib # Website connections
import re # Regular expressions
from time import sleep # To prevent overwhelming the server between connections
from collections import Counter # Keep track of our term counts
from nltk.corpus import stopwords # Filter out stopwords, such as 'the', 'or', 'and'
import pandas as pd # For converting results to a dataframe and bar chart plots
%matplotlib inline
```

```python
def text_cleaner(website):
    try:
        site = urllib.request.urlopen(website).read() # Connect to job posting
    except:
        return
    
    soup_obj = BeautifulSoup(site, 'lxml')
    
    [script.extract() for script in soup_obj(["script", "style"])] # remove script and style elements from BS4 object
    
    text = soup_obj.get_text() # Get the text from soup_obj

    lines = [line.strip() for line in text.splitlines()] #creates list of lines and strips leading/trailing whitespace

    text = ''.join(lines)
    
    text = re.sub(r"([a-z])([A-Z])", r"\1 \2", text) #try to take care of merged words. insert space b/w lower case adjacent to upper
    
    text = re.sub(r'\.([a-zA-Z])', r'. \1', text) #adds space after periods
    
    return text
```

```python
text_cleaner('http://www.builtincolorado.com/job/data-analyst-5')
```
